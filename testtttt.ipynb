```python
dm['Revenue'] = dm.apply(lambda x: x['CostPerItem']*x['NumberOfItemsPurchased'], axis = 1)
```

```python
daily_revenue_generated_by_country = pd.pivot_table(dm[['Country', 'DAY', 'Revenue']], index=['Country'], columns=['DAY'], values=['Revenue'], aggfunc=('sum','mean','min','max')).fillna(0)
daily_revenue_generated_by_country
cols=[]
for x in daily_revenue_generated_by_country.columns:
    c = '_'.join(x).strip('_')
    cols.append(c)
daily_revenue_generated_by_country.columns = cols
daily_revenue_generated_by_country.columns
daily_revenue_generated_by_country.reset_index()
```

```python
monthly_revenue_generated_by_country = pd.pivot_table(dm[['Country', 'MONTH', 'Revenue']], index=['Country'], columns=['MONTH'], values=['Revenue'], aggfunc=('sum','mean','min','max')).fillna(0)
monthly_revenue_generated_by_country
cols=[]
for x in monthly_revenue_generated_by_country.columns:
    c = '_'.join(x).strip('_')
    cols.append(c)
monthly_revenue_generated_by_country.columns = cols
monthly_revenue_generated_by_country.columns
monthly_revenue_generated_by_country = monthly_revenue_generated_by_country.reset_index()
monthly_revenue_generated_by_country
```

```python
yearly_revenue_generated_by_country = pd.pivot_table(dm[['Country', 'YEAR', 'Revenue']], index=['Country'], columns=['YEAR'], values=['Revenue'], aggfunc=('sum','mean','min','max')).fillna(0)
yearly_revenue_generated_by_country
y=[]
y = [str(j) for j in yearly_revenue_generated_by_country.columns]
cols = []
i=0
for x in y:
        c = ''.join(x).strip(',')
        cols.append(c)
cols
for x in cols:
    x = x.replace('(','')
    x = x.replace(')','')
    x = x.replace("'",'')
    x = x.replace('x, ','x in year ')
    x = x.replace('e, ','e computation of ')
    # print(x)
    cols[i] = x
    i=i+1
cols
yearly_revenue_generated_by_country.columns = cols
yearly_revenue_generated_by_country.columns
yearly_revenue_generated_by_country = yearly_revenue_generated_by_country.reset_index()
yearly_revenue_generated_by_country
```

```python
datewise_revenue_generated_by_country = pd.pivot_table(dm[['Country', 'DATE', 'Revenue']], index=['Country'], columns=['DATE'], values=['Revenue'], aggfunc=('sum','mean','min','max')).fillna(0)
datewise_revenue_generated_by_country
y = [str(j) for j in datewise_revenue_generated_by_country.columns]
cols = []
i=0

for x in y:
        c = ''.join(x).strip(',')
        cols.append(c)
cols
for x in cols:
    x = x.replace('(','')
    x = x.replace(')','')
    x = x.replace("'",'')
    x = x.replace('x, ','x in year ')
    x = x.replace('e, ','e computation of ')
    # print(x)
    cols[i] = x
    i=i+1
cols
datewise_revenue_generated_by_country.columns = cols
datewise_revenue_generated_by_country.columns
datewise_revenue_generated_by_country = datewise_revenue_generated_by_country.reset_index()
datewise_revenue_generated_by_country
```

```python
a1 = yearly_revenue_generated_by_country.merge(monthly_revenue_generated_by_country, on = 'Country', how = 'inner')
a2 = datewise_revenue_generated_by_country.merge(daily_revenue_generated_by_country,on = 'Country', how = 'inner')
countrydf = a1.merge(a2,on = 'Country', how = 'inner')
countrydf
```
